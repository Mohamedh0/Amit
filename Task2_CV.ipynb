```python
# Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load an image in grayscale
image = cv2.imread(r'C:\Users\Mohamed\Downloads\WhatsApp Image 2024-11-08 at 12.52.40_11312824.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Apply Gaussian Blur to reduce noise and improve thresholding
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
```

```python
# Apply Otsu's thresholding
_, thresholded_image = cv2.threshold(blurred_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
# Display the original and thresholded images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Otsu Thresholded Image")
plt.imshow(thresholded_image, cmap='gray')
plt.show()
```
